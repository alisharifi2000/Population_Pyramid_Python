```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import imageio
```

```python
data = pd.read_excel('C:\\Users\\a.sharifi\\Desktop\\Untitled Folder 2\\Book3.xlsx')
data.head(10)
```

```python
data = data[['age','year','gender','value']]
data = data.sort_values('age')
data.head(20)
```

```python
order = []
for i in range(0,len(data.age)):
    r = str(data.age.iloc[i])
    r =  r+' years'
    if r not in order:
        order.append(r)
        
print(order)
data.head()
```

```python
for i in range (0,len(data.age)):
    r = str(data.age.iloc[i]) 
    data.age.iloc[i] = r+' years'
data.head(5)
```

```python
counter = 0
for i in range(65,97):
    fig, axes = plt.subplots(ncols=2 ,figsize=(37,20), sharey=True)
    sns.barplot(y="age",x="value", color="darkblue", ax=axes[0],
            data = data[(data.gender == 'male')&(data.year == i)],order = order)
    sns.barplot(y="age",x="value", color="darkred", ax=axes[1],
            data = data[(data.gender == 'female')&(data.year == i)],order = order)
    # Use Matplotlib function to invert the first chart

    # Use Matplotlib function to show tick labels in the middle
    axes[0].yaxis.tick_right()
    
    axes[0].set_title("male")
    axes[1].set_title("female")
    axes[0].set_xlim([0, 1000])
    axes[1].set_xlim([0, 1000])
    axes[0].set(xlabel="Population", ylabel="Age")
    axes[1].set(xlabel="Population", ylabel="")
    axes[0].invert_xaxis()
    fig.suptitle("Population Pyramid (thousands) Iran in 13" + str(i) , size = 25)
    fig.savefig('C:\\Users\\a.sharifi\\Desktop\\Untitled Folder 2\\fig'+ str(counter) +'.png')
    counter = counter + 1
    plt.close()
```

```python
images = []
for i in range(0,counter):
    path = 'C:\\Users\\a.sharifi\\Desktop\\Untitled Folder 2\\fig'+str(i)+'.png'
    print(path)
    images.append(imageio.imread(path))
imageio.mimsave('C:\\Users\\a.sharifi\\Desktop\\Untitled Folder 2\\movie.gif', images, duration=1)
```
